```python
# Read the Excel file
df = pd.read_excel("Magdeburg_PFS_VFS_overlap.xlsx", header=None)

# Convert the DataFrame to a matrix
Coverage = df.values.tolist()
```

```python
# Constants
num_circles = 37

# Circle names dictionary
circle_names = {
    **{f"p{i + 1}": i for i in range(11)},
    **{f"v{i + 1}": i + 11 for i in range(37)}
}

# Set up the problem
prob = pulp.LpProblem("Set_Covering_Problem", pulp.LpMinimize)

# Decision variables
stations = pulp.LpVariable.dicts("Station", range(num_circles), cat="Binary")

# Objective function: minimize the number of stations used
prob += pulp.lpSum([stations[i] for i in range(num_circles)])
```

```python
# Constraints
# Ensure 100% coverage
for i in range(num_circles):
    prob += (
        pulp.lpSum([stations[j] for j in range(num_circles) if Coverage[i][j] == 1]) >= 1
    )
```

```python
# Solve the problem
prob.solve()

# Print the optimal solution
print("Optimal Solution:")
for i in range(num_circles):
    if pulp.value(stations[i]) == 1:
        circle_name = list(circle_names.keys())[list(circle_names.values()).index(i)]
        print(f"{circle_name} - Station Location")

print("Total number of stations used:", int(pulp.value(prob.objective)))
```
